Skewness and Kurtosis

Skewness is a measure of the asymmetry of a data distribution, while kurtosis is a measure of the peakedness of a data distribution. Both measures can help identify underlying patterns in data that can indicate potential outliers or errors in data collection.

What Is Skewness?
Skewness is a statistical measure that assesses the asymmetry of a probability distribution. It quantifies the extent to which the data is skewed or shifted to one side.

Positive skewness indicates a longer tail on the right side of the distribution, while negative skewness indicates a longer tail on the left side. Skewness helps in understanding the shape and outliers in a dataset.

image.png
---
What Is Kurtosis?
Kurtosis is a statistical measure that quantifies the shape of a probability distribution. It provides information about the tails and peakedness of the distribution compared to a normal distribution.

Positive kurtosis indicates heavier tails and a more peaked distribution, while negative kurtosis suggests lighter tails and a flatter distribution. Kurtosis helps in analyzing the characteristics and outliers of a dataset.

The measure of Kurtosis refers to the tailedness of a distribution. Tailedness refers to how often the outliers occur.

Chart visualisation
Display Histogram - Overall Brand Equity

A histogram is a graphical representation of data that is used to show the frequency distribution of a given data set. It is a bar chart that groups a range of data into different bins and displays the number of data points that fall within each bin.

```python
df.OBE.plot.hist() # df.OBE.plot.hist(bins=3)
```

Display a Density Plot

KDE Density Plot: A KDE density plot is a graphical representation of data that displays the probability density of a given data set. It is a smooth curve that shows the estimated probability density of each data point.

```python
df.OBE.plot.kde()
```

```python
df.OBE.plot.kde(ind=[1, 2, 3, 4, 5, 6])
```

Display a Line Chart - OBE

A line plot is a graph that shows a set of data points connected by straight lines. It is used to observe trends and patterns in the data, and can be used to compare multiple data sets on the same graph.

```python
df.OBE.plot()
```

Display a Box Plot - OBE

A box plot is a graphical representation of data that shows the median, quartiles, maximum, and minimum values in a dataset. It is used to compare multiple datasets, identify outliers, and assess overall data distribution and spread.

```python
df.plot.box()
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.box.html
```

```python
df.plot.box(vert=False);
```

```python
df.OBE.plot.box()
```

isplay an Area Plot

An area plot is a two-dimensional graph that displays quantitative data through a filled polygon, which is created by connecting a sequence of points. It is used to visualise the changes in data over a certain period of time, track cumulative totals, and compare the relationship between two or more variables.

```python
df.plot.area()
```

Display a Pie Plot

Pie plots are two-dimensional graphs that display data as a circle divided into sections, with each section representing a proportion of the whole. Pie plots are used to compare the relative size of different categories or groups within a dataset.

```python
series = df.sum()
series.plot.pie(figsize=(4, 4));
```

Scatter plot using pairs of variables (Brand Association vs Overall Brand Equity)

Scatter plots are two-dimensional graphs that display data as a collection of points, with each point representing the values of two variables. The position of the point on the graph indicates how the two variables are related, allowing for easy identification of trends and patterns.

```python
df.plot.scatter(x='BAS',y='OBE')
```

```python
df.plot(x='BAS',y='OBE',kind="scatter")
```
